### 2. Data Cleaning

```python
# View missing values
df.isnull().sum().sort_values(ascending=False)
print(df.isnull().sum())

print("No Null Value detected ")
```

```python
# Check for duplicates and drop if found
duplicate_count = df.duplicated().sum()
print(f"Duplicate rows in dataset: {duplicate_count}")

df = df.drop_duplicates()
```

```python
# Clean 'Skip Rate (%)' column (remove '%' if it's string and convert to float)
if df['Skip Rate (%)'].dtype == 'object':
    df['Skip Rate (%)'] = df['Skip Rate (%)'].str.replace('%', '').astype(float)


numeric_cols = [
    'Monthly Listeners (Millions)',
    'Total Streams (Millions)',
    'Total Hours Streamed (Millions)',
    'Avg Stream Duration (Min)',
    'Streams Last 30 Days (Millions)',
    'Skip Rate (%)'
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)

# Convert 'Release Year' to integer (if needed)
df['Release Year'] = df['Release Year'].astype(int)
df
```

```python
# Optional: rename columns to snake_case for easier code referencing
df.rename(columns={
    'Monthly Listeners (Millions)': 'monthly_listeners_m',
    'Total Streams (Millions)': 'total_streams_m',
    'Total Hours Streamed (Millions)': 'total_hours_streamed_m',
    'Avg Stream Duration (Min)': 'avg_stream_duration_min',
    'Streams Last 30 Days (Millions)': 'streams_last_30_days_m',
    'Skip Rate (%)': 'skip_rate_pct',
    'Release Year': 'release_year',
    'Platform Type': 'platform_type',
    'Country': 'country',
    'Artist': 'artist',
    'Album': 'album',
    'Genre': 'genre'
}, inplace=True)
print(df.head())
print("Rename Success.....")
```

### 4. Univariate Analysis

### 5. Bivariate / Multivariate Analysis

```python
#Relation between Skip Rate & Avg Stream Duration
plt.figure(figsize=(8,6))
sns.scatterplot(x='avg_stream_duration_min', y='skip_rate_pct', data=df, hue='platform_type', palette='Set2')
plt.title('⏱️ Avg Stream Duration vs Skip Rate (%)')
plt.xlabel('Avg Stream Duration (Minutes)')
plt.ylabel('Skip Rate (%)')
plt.grid(True)
plt.show()

print("There's no strong visual relationship between average stream duration and skip rate,")
print("indicating skip behavior isn't clearly driven by how long songs are.")
```

```python
#Correlation Heatmap (Numerical Features)
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('📊 Correlation Heatmap')
plt.show()

print("There is a strong positive correlation between Total Streams and Total Hours Streamed,")
print(" while other variables show very weak or no correlation, especially skip rate and stream duration.")
```

```python
#Platform Type vs Total Streams (Boxplot)
plt.figure(figsize=(8,6))
sns.boxplot(x='platform_type', y='total_streams_m', data=df, palette='pastel')
plt.title('📱 Platform Type vs Total Streams')
plt.xlabel('Platform Type')
plt.ylabel('Total Streams (Millions)')
plt.show()

print("Both Free and Premium users contribute similarly to total streams")
print("but Premium has a slightly higher median and more consistent performance.")
```

### 6. Time-based Trends

```python
#1. Streaming Trends by Release Year

plt.figure(figsize=(12, 6))
sns.lineplot(data=df.groupby('release_year')['total_streams_m'].sum().reset_index(),
             x='release_year', y='total_streams_m', marker='o', linewidth=2.5, color='purple')
plt.title('📆 Total Streams Over Years')
plt.xlabel('Release Year')
plt.ylabel('Total Streams (Millions)')
plt.grid(True)
plt.tight_layout()
plt.show()

print("Total Spotify streams increased overall from 2018 to 2022, peaking in 2022, but saw a sharp decline in 2023.")
```

```python
# 2.Genre popularity trend over the years (Top 5 Genres)
top_genres = df['genre'].value_counts().nlargest(5).index
df_top_genres = df[df['genre'].isin(top_genres)]

plt.figure(figsize=(12, 6))
sns.countplot(data=df_top_genres, x='release_year', hue='genre', palette='Set2')
plt.title('🎶 Top Genres Release Count Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()

print("Hip Hop and EDM releases have shown a steady rise in recent years,")
print("while Rock and Indie genres experienced a noticeable drop in 2023.")
```
